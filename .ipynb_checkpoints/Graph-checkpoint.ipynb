```python
import matplotlib.pyplot as plt
```

```python
file = open("graph_content.txt", "r")
file_lines = file.readlines()
```

```python
all = {}
```

```python
color = [0,0,0,"magenta", "purple", "cyan", "blue", "green", "red", "black", "indigo", "turquoise", "orange", "violet", "brown"]
```

```python
for line in file_lines:

    words = line.split(" ")
    pid_cur = int(words[0])
    tick_no = int(words[1])
    cur_q = int(words[2])

    if pid_cur not in all:
        all[pid_cur] = {}
        all[pid_cur]["x"] = []
        all[pid_cur]["y"] = []

    all[pid_cur]["x"].append(tick_no)
    all[pid_cur]["y"].append(cur_q)

    plt.scatter(tick_no,cur_q, c=color[pid_cur])
```

```python
for pid in all:
    print(color[pid],"\n",all[pid]["x"])
    print(all[pid]["y"])
    temp = str(pid)
    plt.plot(all[pid]["x"], all[pid]["y"],c=color[pid], label=temp)
    plt.legend(loc="lower right")
    plt.savefig('graph.png')
```
